### Средствами библиотеки numpy создать массив со случайными значениями, распределенными по нормальному закону.
### В массиве максимальный элемент заменить на -1000, минимальный на 1000.

```python
import numpy as np
```

```python
# Параметры 
loc = 10 #(Mean) where the peak of the bell exists.
scale = 1 #(Standard Deviation) how flat the graph distribution should be.
size = 5 #The shape of the returned array.
```

```python
# Создание массива
array = np.random.normal(loc, scale, size)
print("Исходный массив:")
print(array)
```

```python
# Замена элементов
initial_max = np.unravel_index(np.argmax(array), array.shape) #np.unravel_index() преобразует линейный индекс в координаты многомерного массива
initial_min = np.unravel_index(np.argmin(array), array.shape)
array[initial_max] = -1000.00
array[initial_min] = 1000.00
print("\nМассив после замены:")
#np.set_printoptions(suppress=True) # Отключение приведения к научной нотации
print(array)
```
